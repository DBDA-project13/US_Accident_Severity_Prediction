```python
import numpy as np # type: ignore
import pandas as pd # type: ignore


# Feature Engineering
def simplify_weather(cond: str) -> str:
    """
    Map raw weather description to a smaller set of categories.
    This reduces the sparsity of Weather_Condition.
    """
    cond = str(cond).lower()

    if "snow" in cond:
        return "snow"
    if "rain" in cond or "storm" in cond or "thunder" in cond or "drizzle" in cond:
        return "rain"
    if "fog" in cond or "mist" in cond or "haze" in cond:
        return "fog"
    if "cloud" in cond or "overcast" in cond:
        return "cloudy"
    if "clear" in cond or "fair" in cond:
        return "clear"
    return "other"

def hour_to_time_bucket(hour):
    """
    Map hour (0–23) to a coarse time-of-day bucket.
    Used later for aggregation.
    """
    if 0 <= hour < 4:
        return "Midnight"
    elif 4 <= hour < 11:
        return "Morning"
    elif 11 <= hour < 15:
        return "Afternoon"
    elif 15 <= hour < 18:
        return "Evening"
    else:
        return "Night"


def transform_distance(df, col='Distance(mi)'):
    """
    Transform the Distance(mi) column:
    1. Cap extreme outliers at 5 miles
    2. Log-transform to reduce skewness
    3. Bin into categories for interpretability
    Returns the dataframe with new features.
    """
    # Cap outliers
    df[f'{col}_capped'] = np.where(df[col] > 5, 5, df[col])
    
    # Log transform
    df[f'{col}_log'] = np.log1p(df[f'{col}_capped'])
    
    # Bin into categories
    bins = [0, 0.01, 0.1, 1, 5]
    labels = ['Zero', 'Very_Short', 'Short', 'Medium']
    df[f'{col}_bin'] = pd.cut(df[f'{col}_capped'], bins=bins, labels=labels, include_lowest=True)
    
    return df


def wind_direction_mapping(direction: str) -> str:
    """
    Map wind direction to a standardized set of values.
    """
    direction = str(direction).upper()
    if direction in ['N', 'NNE', 'NNW']:
        return 'N'
    elif direction in ['S', 'SSE', 'SSW']:
        return 'S'
    elif direction in ['E', 'ENE', 'ESE']:
        return 'E'
    elif direction in ['W', 'WNW', 'WSW']:
        return 'W'
    elif direction in ['NE', 'SE', 'SW', 'NW']:
        return direction
    elif "CALM" in direction:
        return 'CALM'
    elif "VAR" in direction:
        return 'VAR'
    else:
        return 'MISSING'
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType, DoubleType

# Register UDFs
simplify_weather_udf = udf(simplify_weather, StringType())
hour_to_time_bucket_udf = udf(hour_to_time_bucket, StringType())
wind_direction_udf = udf(wind_direction_mapping, StringType())
```

```python
# Weather simplification
df = df.withColumn("Weather_Simple", simplify_weather_udf(col("Weather_Condition")))

# Wind direction mapping
df = df.withColumn("Wind_Direction_Simple", wind_direction_udf(col("Wind_Direction")))

# Extract hour/day/month from Start_Time
df = df.withColumn("Start_Time", to_timestamp(col("Start_Time")))
# df = df.withColumn("hour", hour(col("Start_Time")))
# df = df.withColumn("day", dayofmonth(col("Start_Time")))
# df = df.withColumn("month", month(col("Start_Time")))
df = df.withColumn("hour", hour(col("Start_Time"))) 
df = df.withColumn("day", dayofmonth(col("Start_Time"))) 
df = df.withColumn("month", month(col("Start_Time"))) 
df = df.withColumn("dayofweek", dayofweek(col("Start_Time")))

# Map hour to time bucket
df = df.withColumn("time_bucket", hour_to_time_bucket_udf(col("hour")))
```

```python
from pyspark.sql.functions import log1p, when
from pyspark.ml.feature import Bucketizer

# Cap outliers at 5 miles
df = df.withColumn("Distance_capped", when(col("Distance(mi)") > 5, 5).otherwise(col("Distance(mi)")))

# Log transform
df = df.withColumn("Distance_log", log1p(col("Distance_capped")))

# Bin into categories
splits = [0, 0.01, 0.1, 1, 5]   # same bins as your pandas code
bucketizer = Bucketizer(splits=splits, inputCol="Distance_capped", outputCol="Distance_bin_index")
df = bucketizer.transform(df)

# Map bin index to labels
labels = ['Zero', 'Very_Short', 'Short', 'Medium']
mapping_expr = when(col("Distance_bin_index") == 0, labels[0]) \
    .when(col("Distance_bin_index") == 1, labels[1]) \
    .when(col("Distance_bin_index") == 2, labels[2]) \
    .when(col("Distance_bin_index") == 3, labels[3])
df = df.withColumn("Distance_bin", mapping_expr)
```

```python
from pyspark.sql.functions import col, when, month, dayofweek, hour, lit
import holidays

# --- Rush Hour Indicator ---
df = df.withColumn(
    "is_rush_hour",
    when((col("hour").between(7, 9)) | (col("hour").between(16, 19)), lit(1)).otherwise(lit(0))
)

# --- Season Mapping ---
df = df.withColumn(
    "season",
    when(col("month").isin([12, 1, 2]), "Winter")
    .when(col("month").isin([3, 4, 5]), "Spring")
    .when(col("month").isin([6, 7, 8]), "Summer")
    .when(col("month").isin([9, 10, 11]), "Fall")
)

# --- Day of Week (already extracted) ---
# Spark’s dayofweek returns 1=Sunday … 7=Saturday
# df = df.withColumn("day_of_week", dayofweek(col("Start_Time")))

# --- Holiday Indicator ---
# Use US holidays library
us_holidays = holidays.US()

# Define UDF to check if date is a holiday
from pyspark.sql.functions import udf, to_date
from pyspark.sql.types import IntegerType

def is_holiday(date_str):
    try:
        d = pd.to_datetime(date_str).date()
        return 1 if d in us_holidays else 0
    except:
        return 0

is_holiday_udf = udf(is_holiday, IntegerType())

df = df.withColumn("is_holiday", is_holiday_udf(to_date(col("Start_Time"))))
```
